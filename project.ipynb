```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
data=pd.read_csv('covid_data_2025.csv')
```

```python
state=data['State/UT']
cases=data['Active Cases (Total)']
death=data['Deaths (Total)']

#sort the data firstly zip the data and then sort them
combined=sorted(zip(cases,state,death))
scases,sstate,sdeath=zip(*combined)


plt.LogFormatter(True)
bars=plt.barh(sstate,scases,label="Active Cases",color="#357BD6")

plt.title("Total Covid cases in year 2025 ")
plt.xlabel("death")
plt.ylabel("State and union territories")



for i,bar in enumerate(bars):
    plt.text(bar.get_width()+5,
             bar.get_y()+bar.get_height()/2,
             sdeath[i],
             va='center',fontsize=8,label='deaths')
plt.legend([bars],["Active Cases"])
```

```python
data.sort_values('Deaths (Total)',ascending=False,inplace=True)
```

```python
filtered_df=data[data['Deaths (Total)']>=2]
plt.style.use('ggplot')
states_name=filtered_df['State/UT']
active_cases=filtered_df['Active Cases (Total)']
colors=['#e5ae37','#008fd5','#fc4f30','#6d9042',"#0E5762","#bdbd3f"]
explode=[0,0.02,0,0,0,0]
plt.pie(active_cases,labels=states_name,colors=colors,autopct='%1.1f%%',startangle=25,explode=explode,wedgeprops={'edgecolor':"#444444"})
plt.title("Covid Cases-2025",fontsize=16,color="#0B0000",fontweight='bold',style='oblique')
plt.show()
```
